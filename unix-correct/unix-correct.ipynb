```python
def load_py_dict(path):
    """Load a Python dictionary stored as text."""
    import ast
    with open(path) as f:
        return ast.literal_eval(f.read())
```

```python
add_conf = load_py_dict("spellcorrect/addconfusion.data")
add_conf['th']  # -> number of times 't'+'h' was mistakenly added
```

```python
def load_vocab(words, cutoff=3246):
    """Extract a pruned vocabulary from corpus words."""
    return sorted(set(words))[cutoff:]
```

```python
load_vocab(["a", "the", "actress", "actor", "brilliant"])
# -> ['actress', 'actor', 'brilliant']
```
